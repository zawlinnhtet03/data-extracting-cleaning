```python

df["is_medical_specialist_member"] = df["is_medical_specialist_member"].astype("bool")
df["is_medical_specialist_member"] = df["is_medical_specialist_member"].replace({True: "是", False: "否"})
df[["first_name", "is_medical_specialist_member"]].head()
```

```python
df["is_nuclear_medicine_member"] = df["is_nuclear_medicine_member"].astype("bool")
df["is_nuclear_medicine_member"] = df["is_nuclear_medicine_member"].replace({True: "是", False: "否"})
df[["first_name", "is_nuclear_medicine_member"]].head()
```

```python
df["phone_number"] = df["phone_number"].fillna("0")
# df["phone_number"].head()
df["phone_number"] = df["phone_number"].astype("int")
df.info()
```

```python
df.isnull().any()
```

```python
for col in ["occupation_category", "license_number", "organization", "work_registration_number", "check_in_time", "check_out_time"]:
    df[col] = df[col].fillna("0")
df.head()
```

```python
df = df.rename(columns = {"last_name" : "姓氏",
                          "first_name" : "名字",
                          "is_nuclear_medicine_member" : "是否澳門核醫及分子影像學會會員",
                          "occupation_category" : "職業類別",
                          "license_number" : "執照編號",
                          "is_medical_specialist_member" : "是否澳門醫學専科院土",
                          "organization" : "工作機構",
                          "work_registration_number" : "單位工作登編號",
                          "email" : "電子郵件地址",
                          "phone_number" : "聯絡電話"})
```

```python
df.to_excel("final_data2.0.xlsx", index=False)
```
